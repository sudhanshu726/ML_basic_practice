```python
ca_babies = states_babies.loc[states_babies['State'] == 'CA',:]
```

```python
ca_babies
```

```python
def popular(s):
    """Receives s, a pandas series, containing baby names in order of highest baby count to lower baby count.
    Returns the most popular name in s."""
    return s.iloc[0]
```

```python
ca_grouped = ca_babies.groupby(['Year','Gender']).agg(popular)
```
