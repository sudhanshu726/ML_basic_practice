# Different ways to create Pandas Dataframe
Pandas DataFrame is a 2-dimensional labeled data structure with columns of potentially different types. It is generally the most commonly used pandas object.
Pandas DataFrame can be created in multiple ways. Let’s discuss different ways to create a DataFrame one by one.
Method #1: Creating Pandas DataFrame from lists of lists.

```python
# list of strings
list1 = ['Abi', 'Karthi', 'Nega', 'Vijay','Shalini', 'Pavithra', 'Pranav']
# Calling DataFrame constructor on list
df = pd.DataFrame(list1)
df
```

```python
# list of strings
list1 = ['Abi', 'Karthi', 'Nega', 'Vijay','Shalini', 'Pavithra', 'Pranav']
# Calling DataFrame constructor on list
# with indices and columns specified
df = pd.DataFrame(list1, index =['a', 'b', 'c', 'd', 'e', 'f', 'g'],columns =['Names'])
#Find all the rows based on any condition in a column
df.loc['a']
# select all rows with a condition
```

```python
df.loc['f']
```

```python
#Using zip() for zipping two lists

# list of strings
list1 = ['Abi', 'Karthi', 'Nega', 'Vijay','Shalini', 'Pavithra', 'Pranav']
# list of int
list2 = [11, 22, 33, 44, 55, 66, 77]
# Calling DataFrame constructor after zipping
# both lists, with columns specified
df = pd.DataFrame(list(zip(list1, list2)),columns =['Name', 'val'])
```

```python
df
```

```python
# initialize list of lists
data = [['Jay', 10,10000], ['Manju', 15,25000], ['Sudha', 14,34000]]
# Create the pandas DataFrame
df = pd.DataFrame(data, columns = ['Name', 'Age','Salary'])
# print dataframe.
df.loc[df.Age > 10]
df.loc[(df.Age >= 12) & (df.Salary == 25000)]
```

```python
df.loc[df.Age > 10]
```

```python
#Select only required columns with a condition
df.loc[(df.Age >= 12), ['Name', 'Salary']]
```

```python
# Update the values of a particular column on selected rows
# update a column with condition
df.loc[(df.Age >= 12), ['Salary']] = 56000
df
```

```python
# Update the values of multiple columns on selected rows
# update multiple columns with condition
df.loc[(df.Age >= 10), ['Age', 'Salary']] = [10,50000]
data
```

# Method #2: Creating DataFrame from dict of narray/lists
To create DataFrame from dict of narray/list, all the narray must be of same length. If index is passed then the length index should be equal to the length of arrays. If no index is passed, then by default, index will be range(n) where n is the array length.

```python
# intialise data of lists.
data = {'Name':['Jay', 'Manju', 'Sudha'], 'Age':[20, 21, 19],'Salary':[20000,34000,25000]}
# Create DataFrame
df = pd.DataFrame(data)
# Print the output.
df
```
